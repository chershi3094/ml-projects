```python
# Welcome to Machine Learning with Cherry! >_<
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: study hours vs. marks
hours = np.array([[1], [2], [3], [4], [5]])
marks = np.array([20, 40, 50, 65, 80])

# Train a simple linear regression model
model = LinearRegression()
model.fit(hours, marks)

# Predict marks for 6 hours
predicted = model.predict([[6]])
print(f"Predicted marks for 6 study hours: {predicted[0]:.2f}")

# Plot the data
plt.scatter(hours, marks, color='blue', label='Actual')
plt.plot(hours, model.predict(hours), color='red', label='Prediction')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.legend()
plt.grid(True)
plt.show()
```
